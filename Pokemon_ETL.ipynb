```python
# Dependencies and Setup
# import matplotlib.pyplot as plt
import pandas as pd
# import numpy as np
import requests
# import time

# Import citipy to determine the cities based on latitude and longitude
# from citipy import citipy
from pprint import pprint
```

```python
# Set the API base URL
baseurl = "https://pokeapi.co/api/v2/pokemon/"

# Define an empty list to fetch the poke data 
poke_data = []

poke_requests = requests.get(baseurl).json()

# pprint(poke_requests)
poke_name = []

poke_results = poke_requests["results"]
# print(poke_results)

for i in poke_requests["results"]:
    name = i["name"]
    poke_name.append(name)
print(poke_name)
print(len(poke_name))
```

```python
# testurl = "https://pokeapi.co/api/v2/pokemon/bulbasaur"
# response = requests.get(testurl).json()
# pprint(response)
```

```python
poke_abilities = []
poke_names = []
poke_moves = []
poke_types = []

for j in poke_name:
    url = baseurl + j
    data_requests = requests.get(url).json()
#     pprint(data_requests)

    poke_ability =[]
    poke_move = []
    poke_type = []

    for k in data_requests["abilities"]:
        ability = k["ability"]["name"]
        poke_ability.append(ability)
    poke_abilities.append(poke_ability)
    
    for l in data_requests["forms"]:
        names = l["name"]
        poke_names.append(names)
    
    
    for m in data_requests["moves"]:
        moves = m["move"]["name"]
        poke_move.append(moves)
    poke_moves.append(poke_move)
    
    for n in data_requests["types"]:
        type = n["type"]["name"]
        poke_type.append(type)
    poke_types.append(poke_type)
```

```python
data_df = pd.DataFrame({
    "name": poke_names,
    "abilities": poke_abilities,
    "moves": poke_moves,
    "types": poke_types})

data_df
```
